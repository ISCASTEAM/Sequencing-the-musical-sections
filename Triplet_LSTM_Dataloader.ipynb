```python
import import_ipynb
from torch.utils import data
data_dir = "/media/data/cuixuange/ScrawlMusic/train_data_normalizatin/"
def load_exits_file(train_data,index):
    str_data = train_data[index].split(",")      
    x1_name = data_dir + str_data[0]+".npy"
    x2_name = data_dir + str_data[1]+".npy"
    x3_name = data_dir + str_data[2]+".npy"
    if os.path.isfile(x1_name) and os.path.isfile(x2_name) and os.path.isfile(x3_name): 
        return x1_name,x2_name,x3_name,index
    else:
        index += 6
        return load_exits_file(train_data,index)

class Dataset(data.Dataset):
    def __init__(self, train_data_list, train_label_list):
        'Initialization'
        self.labels = train_label_list
        self.data = train_data_list

    def __len__(self):
        'Denotes the total number of samples'
        return len(self.data)

    def __getitem__(self, index):
        'Generates one sample of data'
        x1_name,x2_name,x3_name,exit_index = load_exits_file(self.data,index)
        X1 = torch.from_numpy(np.expand_dims(np.load(x1_name),axis=2))
        X2 = torch.from_numpy(np.expand_dims(np.load(x2_name),axis=2))
        X3 = torch.from_numpy(np.expand_dims(np.load(x3_name),axis=2))
        X = torch.cat((X1,X2,X3),dim=2)
        #expand the dim of channel 
        X = X.unsqueeze(0) 
        y = torch.tensor(int(self.labels[exit_index]))
        return X, y
```
